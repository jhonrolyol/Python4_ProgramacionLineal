# Programación lineal 

## Naren Castellon

Es una herramienta matemática dedicado a maximizar o minimizar (optimizar) una función lineal, denominado función objetivo de tal forma que las variables de dicha función estén sujetas a una serie de restricciones expresadas mediante un sistema de ecuaciones o inecuaciones también lineales. El método tradicional usado para resolver problema de programación lineal es el $\textbf{método simplex}$

<img src = 'ProgramacionLineal.png' />

El modelado LP/ ILP no es difícil usar un lenguaje de modelado como AMPL, pero puede haber una gran sobrecarga cognitiva al cambiar a un lenguaje completamente diferente, especialmente si lo que estamos haciendo no es solo un análsis ad hoc, sino que es parte de un análisis más amplio, flujo de trabajo continuo. Para aquellos de nosotros que ya pasamos mucho tiempo en Python, sería bueno hacer nuestro trabajo de optmización en el mismo lenguaje que ya estamos usando en cualquier extremo del problema. Afortunadamente, podemos utilizar uno de los muchos paquetes diseñados precisamente para este propósito, tales como: pulp, PyGLPK o PyMathProg.



# Un problema de ejemplo con Python

## Ejemplo. Jans fabrica dos tipos de juguetes de madera: Soldados y Trenes

Un soldado se vende por 27 dólares y utiliza 10 dólares en materias primas. Cada soldado que se fabrica aumenta la mano de obra variable de Jans y los costos generales en 14 dólares. Un tren se vende por 21 dólares y utiliza 9 dólares en materias primas. Cada tren contruido aumenta la mano de obra variable y los costos generales de Jans en 10 dólares. La fabricación de soldados y trenes de madera requiere dos tipos de mano de obra calificada: Carpinteria y acabado.
Un soldado requiere 2 horas de trabajo de acabado y 1 hora de trabajo de carpinteria. Un tren requiere 1 hora de trabajo de acabado y 1 hora de trabajo de carpinteria. Cada semana Jans puede obterner toda la materia prima necesaria pero solo 100 horas de acabado y 80 horas de carpinteria. La demanda de trenes es ilimitda, pero como máximo se compran 40 soldados cada semana. Jans quiere maximizar las ganancias semanales (ingresos - costos)

Formule un modelo matemático de la situación de Jans que pueda usarse para maximizar las ganancias semanales de Jans.

## Función objetivo

$s$: Npumero de unidades de soldados 

$t$: Número de unidades de 

$$
F.O = 3s + 2t
$$

## Sujeta a:

### Restricción 1

$$
s + t \leq 80
$$

### Restricción 2

$$
2s + t \leq 100
$$

### Restricción 3

$$
s \leq 40
$$

### Restricción de no negatividad

$$
s, t \geq 0
$$

Primero importaremos los paquetes necesarios y crearemos el pulpobjetivo LP con el que trabajaremos



```python
# pip install pulp
```

```python
import numpy as np
import pulp
```

```python
# Creamos el objeto del LP, para un problema de maximización
Prob = pulp.LpProblem("Jans", pulp.LpMaximize)
```

```python
# Creamos las varibales de decisión
Soldado = pulp.LpVariable('Soldados', lowBound = 0, cat = 'Integer')
Tren = pulp.LpVariable('Tren',lowBound = 0, cat = 'Integer')
```

Ahora agregamos la función objetivo:

```python
# Modelo de producción de costo semanal
CostosMateriaPrima = 10*Soldado + 9*Tren
VariableCosts = 14*Soldado + 10*Tren
```

```python
# Ingreso por venta de los juguetes
Ventas = 27*Soldado + 21*Tren
```

```python
# Función objetivo a maximizar 
Profit = Ventas - (CostosMateriaPrima + VariableCosts)
Prob += Profit # Acá es realmente donde agregamos la función objetivo
```

```python
Prob
```

Ahora creamos las restricciones

```python
# Restricciones por fabricación de juguetes por horas
CarpinteriaHora = Soldado + Tren
Prob += (CarpinteriaHora <= 80)

AcabadoHora = 2*Soldado + Tren 
Prob += (AcabadoHora <= 100)
```

```python
# Demanda por soldados 
Prob +=  (Soldado <= 40)
```

Imprimamos el problema y asegurémonos de tener todo:

```python
print(Prob)
```

Ya hemos formulado nuestro problema, ahora buscamos una manera de resolver nuestro LP

```python
# Resolviendo en problema LP usando Solver
OptimizationResult = Prob.solve()
```

```python
# Buscamos la solución óptima
assert OptimizationResult == pulp.LpStatusOptimal
```

```python
# Visualizamos resultados
for var in (Soldado, Tren):
    print('Número de semanas óptimas {} a producir: {:1.0f}'.format(var.name, var.value()))
```

## Comprovando la respuesta

También podemos visualizar este problema y verificar que la respuesta que hemos encontrado tiene sentido.
Aquí están las limitaciones:

$$
s + t \leq 80
$$

$$
2s + t \leq 100
$$

$$
0 \leq s \leq 80
$$

$$
t \geq 0
$$

$$
s,t \geq 0
$$

Podemos escribir los que tienen ambos $s$ y $t$ de la siguiente manera para que se puedan trazar como líneas en un gráfico 2D con un $s$-eje y un $t$-eje

$$
t \leq 80 - s
$$

$$
t \leq 100 - 2s
$$

Grafiquemos todas estas desigualdades simultáneas.

```python
%matplotlib inline 
from matplotlib import pyplot as plt
from matplotlib.path import Path
from matplotlib.patches import PathPatch

import seaborn as sns
#sns.set_color_palette("set1")
```

```python

# Creamos nuestro gráfico con nuestras restricciones 
fig, ax = plt.subplots(figsize= (8,8))
s = np.linspace(0,100)

# Restricción por carpinteria: Tren <= 80 - Soldado
plt.plot(s, 80 - s, lw = 3, label = 'Carpinteria')
plt.fill_between(s,0,80 -s, alpha = 0.1)


# Restricción por acabado: Tren  <= 100 - 2*Soldado
plt.plot(s, 100 - 2*s, lw = 3, label = 'Acabado')
plt.fill_between(s,0,100 -2*s, alpha = 0.1)


# Restricción de la demanda: Soldado <= 40
plt.plot(40*np.ones_like(s), s, lw = 3, label = 'Demanda')
plt.fill_betweenx(s,0,40, alpha = 0.1)

# Agregamos restricciones de no negatividad 
plt.plot(np.zeros_like(s),s,lw = 3, label = 't non - negative')
plt.plot(s,np.zeros_like(s),lw = 3, label = 's non - negative' )

# Sobrear la región factible 
path = Path([
    (0., 0.),
    (0., 80.),
    (20.,60.),
    (40.,20.),
    (40.,0.),
    (0.,0.),    
])
patch = PathPatch(path, label = 'Región Factible', alpha = 0.5)
ax.add_patch(patch)

# Etiquetas y relleno 
plt.xlabel('Solado', fontsize = 16)
plt.ylabel('Tren',fontsize = 16)
plt.xlim(-0.5, 100)
plt.ylim(-0.5,100)
plt.legend(fontsize = 14)
plt.show()  

```

El área resaltada muestra el conjunto de decisiones sobre $s$ y $t$ que satisfacen todas las restricciones, esta es la región 
factible. Por inspección, esto es claramente un politopo convexo.

Una posible queja sobre este gráfico esque es un poco engañoso. La región sólida factible parace sugerir una área continua de posibles opciones para $s$ y $t$. Pero este no es el caso. Una última restricción es que las variables de decisión deben ser números entéros. Veamos si podemos mejorar el trama:



```python
fig, ax = plt.subplots(figsize=(9,8))
s = np.linspace(0,100)

# Graficamos las restricciones 
plt.plot(s,80-s,lw = 3,label = 'Carpinteria')
plt.plot(s,100-2*s,lw = 3,label = 'Acabado')
plt.plot(40*np.ones_like(s),s,lw = 3, label = 'Demanda')
plt.plot(np.zeros_like(s),s,lw = 3, label = 't non - negative')
plt.plot(s,np.zeros_like(s),lw = 3, label = 't non - negative')


# Gráfico de los posibles pares (s,t) pairs
pairs = [(s,t) for s in np.arange(101)
               for t in np.arange(101)
               if (s + t) <= 80
               and (2*s + t) <= 100
               and s <= 40]

# División de nuestras variables
ss, ts = np.hsplit(np.array(pairs),2)

# Calcular la función objetivo en cada par 
z = 3*ss + 2*ts # Función objetivo

# Graficar los resultados
plt.scatter(ss, ts, c = z, cmap = 'jet',label = 'Ganancia', zorder = 3)
# Etiquetas y sombreado
cb = plt.colorbar()
cb.set_label('Ganancia', fontsize = 14)
plt.xlabel('Soldado', fontsize = 16)
plt.ylabel('Tren', fontsize = 16)
plt.xlim(-0.5, 100)
plt.ylim(-0.5, 100)
plt.legend(fontsize = 14)
plt.show()
```

Esta gráfica también muestra que el punto óptimo está en un vértice como se esperaba, mientras que las franjas de color 
claramente paralelas demuestran una frontera de decisión lineal para diferentes valores de la función objetivo.

Una vez hecho este trabajo, podemos decirlo con confianza a Jans cuántos de cada juguete vender para maximizar sus ganancias, e incluso mostrarle cuánto podría ganar en cualquiera de los otros posibles puntos de decisión.
